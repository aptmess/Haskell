## Рекурсия в лямбда-исчислении

Как можно рекурсию выразить в лямбда-исчислении.

```haskell
fac = λn.(if (= n 0) 1 (* n (fac (- n 1))))
```

Попытаемся дать имя лямбда-выражению. Это лямбда выражение от $n$ - проверяющая с помощью функции `if` верно ли, что $n$ равно нулю, иначе вызывает от $n-1$ - классическое определение факториала - именованное выражение.

Прежде всего давайте построим другое выражение, не представляющее факториал - рекурсивное обращение сделаем аргументом другой функции.

`FB  = λf.λn.(if (= n 0) 1 (* n (f (- n 1))))`

Найти `fac` такой, что `fac = FB fac`

Будем говорить про неподвижную точку функции - нельзя ли написать функцию для применения к FB. Если такая функция к вычислению неподвижной точки существует, То это будет тот самый инструмент, который нам нужен.



если применим $Y$ к $F$, то это будет то же самое, что и функция $F$ к $Y \ F$

Такая функция называется **Y-комбинатор** - комбинатор фиксированной точки - если мы найдем такую функцию, то мы найдем способ выражать рекурсию.

## Чистое лямбда-исчисление

Чистое лямбда-исчисление – это исчисление без констант и встроенных функций. Соответственно, отсутствует и $\delta$-редукция; есть только $\lambda$-выражения и $\beta$-редукция.

Покажем, что выразительная сила чистого лямбда-исчисления не меньше, чем у традиционного функционального языка программирования.
